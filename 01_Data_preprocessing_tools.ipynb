# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
# For features we use x which is a matrix
# iloc used to locate indices
# iloc[row range, col range]
x = dataset.iloc[:,:-1].values
# For depended variable vector
y = dataset.iloc[:,-1].values  
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan, strategy = 'mean')
#fit method  -> connect imputer to methods of features
#then replacement is done by using transform method
imputer.fit(x[:,1:3])  
x[:,1:3] = imputer.transform(x[:,1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
#One hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(), [0])], remainder='passthrough') # remainder = 'passthrough' is use so as to retain salary and age column .
# However, if remainder column is not specified then new feature matrix will only contain 3 column and no salary and age column.
x = np.array(ct.fit_transform(x))
```

```
print(x)
```

### Encoding the Dependent Variable

```
#Label Encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2,random_state =1)
```

```
print(x_train)
```

```
print(x_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# fit method here will only calculate standard deviation and mean of the values
# transform will apply the formula of standardization to the column
# fit_transform will do both the above work at the same time
x_train[:,3:] = sc.fit_transform(x_train[:,3:])
# transform the features of the test set 
# if we use new sc.fit_transform() method then we will get a new scaler which we don't want.
# we need to use the same scaler to be applied in the test set
x_test[:,3:] = sc.transform(x_test[:,3:])
```
